- Is there any duplicate Record in the dataset? If yes, then remove the duplicate records.

```python
df.duplicated().sum()
```

```python
# As we saw that there are 2 duplicated rows in this dataset
#  So now we remove this duplicated data

df.drop_duplicates(inplace=True)
```

- Is there any null values present in the dataset? If Yes, Remove the null values.

```python
df.isnull().sum()
```

```python
df.Director = df.Director.fillna('UNKNOWN')
```

```python
df.cast = df.Cast.fillna('UNKNOWN')
```

```python
df.Country = df.Country.fillna('UNKNOWN')
```

```python
df.Release_Date = df.Release_Date.fillna(0)
```

```python
df.Rating = df.Rating.fillna('UNKNOWN')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

- In which year Highest number of the TV shows & Movies were released? Show with Bar Graph.

```python
df['Data_and_year'].dt.year.value_counts()
```
